```python
import numpy as np

from latent_analysis import LatentAnalysis

LA = LatentAnalysis()
```

#### Load muscle-wingkin dataset

Load the muscle and wing motion dataset and apply outlier-rejection.

```python
dataset_loc  = LA.working_dir / 'dataset'
dataset_file = dataset_loc / 'main_muscle_and_wing_data.h5'

LA.load_dataset(dataset_file)

outliers = np.array([
    [18,0,100],
    [230,0,1000],
    [231,0,1000],
    [232,0,1000],
    [261,0,50],
    [338,0,1000],
    [345,100,150],
    [346,0,1000],
    [354,0,120]])

LA.create_dataset(outliers)
```

#### Build latent network

Construct the latent network consisting of: an encoder, a muscle-decoder and a wingkin-decoder.

```python
N_filters = 64
N_latent = 5

LA.build_network(N_filters,N_latent)
```

#### Train latent network or load weights

The default is to load the network weights located in the 'weights' folder. In case you want to train the latent network on the dataset, you can uncomment the training-section and comment the load-weights-section.

```python
weight_loc = LA.working_dir / 'weights'

# train network:
#LA.train_network(N_filters,N_latent,weight_loc)

# load weights:
LA.load_network(N_filters,N_latent,weight_loc)
```

####  Plot latent modes

The value of each latent variable varies between $-3 \sigma$ and $+3 \sigma$, while the other variables are kept at $0$. By using the muscle-decoder and wingkin-decoder, we can reconstruct the muscle activity patterns (per sclerite) and resulting wing motion as a function of the latent variable. The muscle activity is plotted against the latent variable, while the wing kinematics use a color-code to indicate the latent value: the darkest blue corresponding to $-3 \sigma$ and the darkest red to $+3 \sigma$.

```python
LA.create_plot_locations()

plot_loc = LA.working_dir / 'plots'

LA.plot_latent_modes(N_latent,plot_loc)
```

#### Compute angle-of-attack, wingtip velocity, lift and drag forces

With the latent wing kinematics known, we can use the quasi-steady model and the Newton-Euler equations to compute the forces on the wing. We postulate that a compliant cross-section at the wing root functions as a spring and forces on the wing can thereby (passively) alter the stroke and deviation angles. To illustrate the passive effect of this spring on wing motion we plot the angle-of-attack, wingtip velocity and the lift and drag forces as a function of each latent parameter. 

```python
LA.plot_LD_latent_modes(N_latent,plot_loc)
```
